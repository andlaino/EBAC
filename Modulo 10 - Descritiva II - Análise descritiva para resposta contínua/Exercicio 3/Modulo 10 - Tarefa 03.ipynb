````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
````

````python
df = pd.read_csv('dados/Profissao- Cientista de Dados_M10_support material.csv')

df = df.drop(columns='Unnamed: 0')
df.drop_duplicates(inplace=True)

df.head()
````

1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
df['data_ref'] = pd.to_datetime(arg=df['data_ref'])
# Verificando variáveis qualitativas:

print(df.select_dtypes(exclude=['datetime', 'number']).columns)
# ou
print(df.select_dtypes(include=['object', 'boolean']).columns)
````

````python
qualitativas = df.select_dtypes(include=['object', 'boolean']).columns

print(df[qualitativas].info())
df[qualitativas]
````

````python
plt.rc('figure', figsize=(12,6))

for col in qualitativas:
    plt.figure(figsize=(12,6))  # Criar uma nova figura para cada gráfico
    ax = sns.countplot(x='data_ref', hue=col, data=df)
    
    tick_labels = df['data_ref'].map(lambda x: x.strftime('%b/%Y')).unique()
    ax.set_xticks(range(len(tick_labels)))  # Define as posições dos ticks
    ax.set_xticklabels(labels=tick_labels, rotation=45)  # Define os rótulos corretamente
    
    ax.legend(bbox_to_anchor=(1, 0.5), loc='center left', title=f"'{col}'")
    plt.show()
````

Todos os gráficos apresentados, uma das variáveis qualitativas que se destaca das demais é a escolaridade. Ao longo do tempo, a partir de setembro de 2015, houve um aumento significativo de dados com a escolaridade de ensino superior completo, o que resultou em uma queda na quantidade de dados com a escolaridade secundária.

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
plt.rc('figure', figsize=(16, 4))

for col in qualitativas:
    fig, axes = plt.subplots(1, 2, figsize=(16, 4))
    fig.subplots_adjust(wspace=0.6)

    tick_labels = df['data_ref'].map(lambda x: x.strftime('%b/%Y')).unique()
    tick_positions = range(len(tick_labels))  # Define posições fixas para os ticks

    # Barras empilhadas:
    df_crosstab = pd.crosstab(index=df['data_ref'], 
                              columns=df[col], 
                              normalize='index')
    
    ax0 = df_crosstab.plot.bar(stacked=True, ax=axes[0])
    ax0.set_xticks(tick_positions)  # Define os ticks corretamente
    ax0.set_xticklabels(labels=tick_labels, rotation=45)
    axes[0].legend(bbox_to_anchor=(1, 0.5), loc='center left', title=f"'{col}'")

    # Perfis médios no tempo:
    ax1 = sns.pointplot(x='data_ref', y='renda', hue=col, data=df, dodge=True, errorbar=('ci', 95), ax=axes[1])
    ax1.set_xticks(tick_positions)  # Define os ticks corretamente
    ax1.set_xticklabels(labels=tick_labels, rotation=45)
    axes[1].legend(bbox_to_anchor=(1, 0.5), loc='center left', title=f"'{col}'")

    plt.show()
````

Como descrito na resposta do exercício anterior, houve uma alteração na variável escolaridade, o que resultou em uma diferença em praticamente metade do período analisado. Exceto por esta variável mencionada, acredito que todas as outras variáveis qualitativas possam ser utilizadas em um modelo devido à estabilidade de distribuição ao longo do tempo.
