```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df_base = pd.read_csv('dados/sinasc.csv')

print (df_base.info())

df_base
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
df_base.loc[(df_base['APGAR5'] >= 0) & 
              (df_base['APGAR5'] <= 3), 
              'APGAR5_categorizado'] = 'asfixia grave'

df_base.loc[(df_base['APGAR5'] >= 4) & 
              (df_base['APGAR5'] <= 5), 
              'APGAR5_categorizado'] = 'asfixia moderada'

df_base.loc[(df_base['APGAR5'] >= 6) & 
              (df_base['APGAR5'] <= 7), 
              'APGAR5_categorizado'] = 'asfixia leve'

df_base.loc[(df_base['APGAR5'] >= 8) & 
              (df_base['APGAR5'] <= 10), 
              'APGAR5_categorizado'] = 'normal'

df_base['APGAR5_categorizado'].value_counts()
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
freq = pd.crosstab(df_base['APGAR5'], df_base['SEXO'])
print(freq)
print(freq.sum(1))

freq_pctlin = freq.div(freq.sum(axis = 1), axis = 0)
print(freq_pctlin)

freq_pctlin.plot.bar()

plt.show()
```

```python
sns.catplot(x='GESTACAO', 
            y='APGAR5',
            hue='SEXO', 
            data=df_base, 
            order=['Menos de 22 semanas', 
                   '22 a 27 semanas', 
                   '28 a 31 semanas', 
                   '32 a 36 semanas', 
                   '37 a 41 semanas', 
                   '42 semanas e mais'], 
            kind='bar')
plt.xticks(rotation=45)
plt.xlabel(xlabel='Tempo de gestação')
plt.ylabel(ylabel="Média de 'APGAR5'")

plt.show()
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
tabela_cruzada = pd.crosstab(index=df_base['GESTACAO'], 
                             columns=df_base['APGAR5_categorizado'])

tabela_cruzada = tabela_cruzada.reindex(
    index=['Menos de 22 semanas', 
           '22 a 27 semanas', 
           '28 a 31 semanas', 
           '32 a 36 semanas', 
           '37 a 41 semanas', 
           '42 semanas e mais'], 
    columns=['asfixia grave', 
             'asfixia moderada', 
             'asfixia leve', 
             'normal'])

tabela_cruzada
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
pct_tabela_cruzada = tabela_cruzada.div(tabela_cruzada.sum(axis=1), axis=0)

print(pct_tabela_cruzada.sum(axis=1))

pct_tabela_cruzada
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
plt.rc('figure', figsize=(15,6))

pct_tabela_cruzada.plot.bar(color=['red', 'orange', 'yellow', 'green'])
plt.xticks(rotation=45)
plt.xlabel(xlabel='Tempo de gestação')
plt.legend(title='Apgar no quinto minuto')

plt.show()
```
