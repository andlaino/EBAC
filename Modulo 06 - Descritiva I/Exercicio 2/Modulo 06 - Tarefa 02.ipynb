```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df_base = pd.read_csv('sinasc.csv')

print (df_base.info())

df_base
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
plt.rc('figure', figsize=(8,5))

fig, axes = plt.subplots(1,1)

df_base[['IDADEMAE']].boxplot()
sns.boxplot(y="IDADEMAE", data=df_base[df_base['IDADEMAE']>1])

plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
plt.rc('figure', figsize=(8,5))

fig, axes = plt.subplots(1,1)

df_base[['PESO']].boxplot()
sns.boxplot(y='PESO', data=df_base[df_base['PESO']>1])

plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
plt.rc('figure', figsize=(8,5))

fig, axes = plt.subplots(1,1)

df_base['IDADEMAE'].plot.hist(bins = 20)

sns.displot(data = df_base[df_base['IDADEMAE']>1]
            , alpha = .50 
            , x = 'IDADEMAE'
            , bins = 20
            , element = 'step'
            , kde = True
            )

plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
plt.rc('figure', figsize=(8,5))

fig, axes = plt.subplots(1,1)

df_base['PESO'].plot.hist(bins = 20)

#afim de executar outro tipo de grafico
sns.displot(data = df_base[df_base['PESO']>1]
            , alpha = .50 
            , x = 'PESO'
            , bins = 20
            )

plt.show()
```
