```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_base = pd.read_csv('dados/sinasc.csv')

print (df_base.info())

df_base
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
df_base['APGAR1'].value_counts().sort_index().plot.bar()
plt.title(label="'APGAR1'")
plt.xlabel(xlabel='Apgar no primeiro minuto')
plt.tick_params(axis='x', rotation=0)
plt.show()
```

```python
df_base['APGAR5'].value_counts().sort_index().plot.bar()
plt.title(label="'APGAR5'")
plt.xlabel(xlabel='Apgar no quinto minuto')
plt.tick_params(axis='x', rotation=0)
plt.show()
```

```python
df_base['SEXO'].value_counts().plot.bar()
plt.title(label="'SEXO'")
plt.tick_params(axis='x', rotation=0)
plt.show()
```

```python
df_base['GESTACAO'].value_counts().reindex(
    ['Menos de 22 semanas', 
     '22 a 27 semanas', 
     '28 a 31 semanas', 
     '32 a 36 semanas', 
     '37 a 41 semanas', 
     '42 semanas e mais']).plot.bar()
plt.title(label="'GESTACAO'")
plt.xlabel(xlabel='Semanas de gestação')
plt.tick_params(axis='x', rotation=45)
plt.show()
```

```python
df_base['ESTCIVMAE'].value_counts().sort_index().plot.bar()
plt.title(label="'ESTCIVMAE'")
plt.xlabel(xlabel='Estado civil: \n\n1: Solteira \n2: Casada \n3: Viuva \n4: Separado judicialmente/Divorciado \n5: União consensual \n9: Ignorado')
plt.tick_params(axis='x', rotation=0)
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
# Exibindo legenda explicativa
print('Apgar no quinto minuto: \n0 a 10 \n')
print('Número de consultas de pré-natal: \n1: Nenhuma \n2: de 1 a 3 \n3: de 4 a 6 \n4: 7 e mais \n9: Ignorado')

# Configurando o tamanho da figura
plt.rc('figure', figsize=(5, 5))

# Criando um gráfico de barras colorido corretamente
ax = sns.barplot(
    data=df_base, 
    x='APGAR5', 
    y='CONSULTAS',
    hue='APGAR5',  # Define a cor de acordo com os valores da variável x
    palette="viridis",  
    legend=False  # Remove a legenda desnecessária
)

# Ajustando os ticks do eixo Y
ax.set_yticks(ticks=[1, 2, 3, 4])

# Configurando o rótulo do eixo X
plt.xlabel("'APGAR5': Apgar no quinto minuto")

# Exibindo o gráfico
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
# Configurando o tamanho da figura
plt.rc('figure', figsize=(16, 16))
fig, axes = plt.subplots(2, 2)
fig.subplots_adjust(hspace=0.4)

# Gráfico 1 - Sexo (cores azuis)
ax1 = sns.countplot(x='SEXO', data=df_base, hue='SEXO', palette="Blues", ax=axes[0, 0], legend=False)

# Gráfico 2 - Semanas de gestação (cores viridis)
ax2 = sns.countplot(
    x='GESTACAO', 
    data=df_base, 
    hue='GESTACAO',
    order=['Menos de 22 semanas', 
           '22 a 27 semanas', 
           '28 a 31 semanas', 
           '32 a 36 semanas', 
           '37 a 41 semanas', 
           '42 semanas e mais'], 
    palette="viridis",
    ax=axes[0, 1], 
    legend=False
)
ax2.set_xticks(range(len(ax2.get_xticklabels())))  # Definir posições fixas
ax2.set_xticklabels(labels=[tick.get_text() for tick in ax2.get_xticklabels()], rotation=45)
ax2.set_xlabel(xlabel='Semanas de gestação')

# Gráfico 3 - Estado Civil (cores pastéis)
ax3 = sns.countplot(x='ESTCIVMAE', data=df_base, hue='ESTCIVMAE', palette="pastel", ax=axes[1, 0], legend=False)
ax3.set_xticks(range(len(ax3.get_xticklabels())))  # Definir posições fixas
ax3.set_xticklabels(labels=['1: Solteira', 
                            '2: Casada', 
                            '3: Viúva', 
                            '4: Separado judicialmente\nDivorciado', 
                            '5: União consensual',
                            '9: Ignorado'], rotation=45)
ax3.set_xlabel(xlabel='Estado civil')

# Gráfico 4 - Consultas Pré-Natal (cores plasma)
ax4 = sns.countplot(x='CONSULTAS', data=df_base, hue='CONSULTAS', palette="plasma", ax=axes[1, 1], legend=False)
ax4.set_xticks(range(len(ax4.get_xticklabels())))  # Definir posições fixas
ax4.set_xticklabels(labels=['1: Nenhuma', 
                            '2: de 1 a 3', 
                            '3: de 4 a 6', 
                            '4: 7 e mais', 
                            '9: Ignorado'], rotation=45)
ax4.set_xlabel(xlabel='Número de consultas de pré-natal')

# Exibir os gráficos
plt.show()
```
